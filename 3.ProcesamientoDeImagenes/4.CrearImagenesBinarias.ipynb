```python
import matplotlib.pyplot as plt
import cv2
```

```python
#Convertiamos a una escala de grises
imagenOriginal=cv2.imread("D:\CursoOpenCV\Imagenes\Yukino.jpg",0)
plt.imshow(imagenOriginal, cmap='gray')
```

```python
#Convetiamos a una imagen binaria
#threshold(IMAGEN,VALORMNIMO,VALORMAXIMO,cv2.THRESH_BINARY)
mitadImagen,imagenOriginal=cv2.threshold(imagenOriginal,255/2,255,cv2.THRESH_BINARY)
plt.imshow(imagenOriginal, cmap='gray')
```
