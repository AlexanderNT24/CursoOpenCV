```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
imagenOriginal=cv2.imread("D:\CursoOpenCV\Imagenes\Yukino.jpg")
imagen2=cv2.cvtColor(imagenOriginal,cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
imagen2.shape
```

```python
#Eje X: Valor de cada pixel
#Eje Y: Frecuencia con la que apararece

histo=cv2.calcHist([imagen2],[0],None,[256],[0,256])
plt.plot(histo)
```

```python
colores=('b','g','r')
#Un bucle donde recorremos vada valor de color
for i,color in enumerate(colores):
    histo=cv2.calcHist([imagen2],[i],None,[256],[0,256])
    plt.plot(histo)
    plt.xlim([0,256])
```
