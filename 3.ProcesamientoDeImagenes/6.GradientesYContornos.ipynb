```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
imagenOriginal=cv2.imread("D:\CursoOpenCV\Imagenes\sudoku.jpg",0)
plt.imshow(imagenOriginal, cmap='gray')
```

```python
#Sobel(IMAGEN,cv2.CV_64F,y,x,ksize=5)
sobelx=cv2.Sobel(imagenOriginal,cv2.CV_64F,1,0)
plt.imshow(sobelx, cmap='gray')
```

```python
#Sobel(IMAGEN,cv2.CV_64F,y,x,ksize=5)
sobely=cv2.Sobel(imagenOriginal,cv2.CV_64F,0,1)
plt.imshow(sobely, cmap='gray')
```

```python
imagenUnida=cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
plt.imshow(imagenUnida, cmap='gray')
```

```python
laplacian=cv2.Laplacian(imagenOriginal,cv2.CV_64F)
plt.imshow(laplacian, cmap='gray')
```
