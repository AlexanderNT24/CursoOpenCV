```python
import matplotlib.pyplot as plt
import cv2
```

```python
#Opencv usa el formato BGR y el matplotlib usa RGB
imagenOriginal=cv2.imread("D:\CursoOpenCV\Imagenes\Yukino.jpg")
plt.imshow(imagenOriginal)
```

```python
imagenY2=cv2.cvtColor(imagenOriginal,cv2.COLOR_BGR2RGB)
plt.imshow(imagenY2)
```

```python
imagenOriginal2=cv2.imread("D:\CursoOpenCV\Imagenes\Hikki.jpg")
plt.imshow(imagenOriginal2)
```

```python
imagenH2=cv2.cvtColor(imagenOriginal2,cv2.COLOR_BGR2RGB)
plt.imshow(imagenH2)
```

```python
#CAMBIAMOS EL TAMAO PARA UNIFICARLOS
imagenY2=cv2.resize(imagenY2,(1200,1200))
imagenH2=cv2.resize(imagenH2,(1200,1200))
plt.imshow(imagenY2)
```

```python
plt.imshow(imagenH2)
```

```python
#cv2.addWeighted(src1=IMAGEN,alpha=PESO,src2=IMAGEN,beta=PESO)
imagenUnida=cv2.addWeighted(src1=imagenY2,alpha=0.5,src2=imagenH2,beta=0.5,gamma=0)
plt.imshow(imagenUnida)
```
