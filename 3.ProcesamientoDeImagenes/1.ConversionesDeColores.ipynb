```python
'''
Espacios de colores: RGB (RED GREEN BLUE) BGR(BLUE GREEN RED)

HSV (MATIZ SATURACION Y VALOR)

HSL (MATIZ SATURACION Y LUMINOSIDAD)
'''
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
#Opencv usa el formato BGR y el matplotlib usa RGB
imagenOriginal=cv2.imread("D:\CursoOpenCV\Imagenes\Yukino.jpg")
plt.imshow(imagenOriginal)
```

```python
imagen2=cv2.cvtColor(imagenOriginal,cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
imagen=cv2.cvtColor(imagenOriginal,cv2.COLOR_BGR2HSV)
plt.imshow(imagen)
```

```python
imagen=cv2.cvtColor(imagenOriginal,cv2.COLOR_BGR2HLS)
plt.imshow(imagen)
```
