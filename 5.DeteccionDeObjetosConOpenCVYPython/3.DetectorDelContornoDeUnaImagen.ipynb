```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
imagen=cv2.imread("D:\CursoOpenCV\Imagenes\YukinoCara.jpg")
```

```python
plt.imshow(imagen)
```

```python
#https://docs.opencv.org/3.4/da/d22/tutorial_py_canny.html
#https://java2blog.com/cv2-canny-python/
#Con el metodo cv2.Canny(imagen,ValorDeUmbralAltoDelGradienteDeIntensidad,ValorDeUmbralBajoDelGradienteDeIntensidad)
contorno=cv2.Canny(image=imagen,threshold1=127,threshold2=127)
plt.imshow(contorno)
```

```python
#Con el metodo cv2.blur(imagen,ksize=(5,5)) suavizamos la imagen 
#cv2.blur()El mtodo se utiliza para desenfocar una imagen utilizando el filtro de cuadro normalizado
#https://www.geeksforgeeks.org/python-opencv-cv2-blur-method/
imagen=cv2.blur(imagen,ksize=(5,5))
```

```python
contorno=cv2.Canny(image=imagen,threshold1=127,threshold2=127)
plt.imshow(contorno)
```
