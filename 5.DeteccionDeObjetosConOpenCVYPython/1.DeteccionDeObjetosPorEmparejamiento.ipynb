```python
import matplotlib.pyplot as plt
import cv2
```

```python
#Leo la imagen 
imagen=cv2.imread("D:\CursoOpenCV\Imagenes\Yukino.jpg")
imagen=cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
```

```python
#Vamos a identificar la cara dentro de la imagen con un rectangulo
imagen2=cv2.imread("D:\CursoOpenCV\Imagenes\YukinoCara.jpg")
imagen2=cv2.cvtColor(imagen2,cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
#Identificamos la cara mediante el metodo
#El metodo es cv2.TM_CCOEFF
metodo=cv2.TM_CCOEFF
#Con el metodo cv2.matchTemplate(imagen,imagen2,metodo) pasamos como parametro las dos imagenes a comparar y el metodo
resultado=cv2.matchTemplate(imagen,imagen2,metodo)
#El resultado es una imagen como un mapa de calor donde nos muestra el punto
#De la imagen donde hay mayor coincidencia comparando en la imagen1 la imagen2
plt.imshow(resultado)
```

```python
#Vamos a usar el mapa de calor para identificar el punto donde esta la cara
#Usamos la funcion minMaxLoc(resultado) la cual nos devuelve valores minimos maximos y sus posiciones (coordenadas)
valorMinimo,valorMaximo,poscionMinima,poscionMaxima=cv2.minMaxLoc(resultado)
```

```python
#Calculamos el tamao de la imagen2 donde esta la cara
imagen2.shape
```

```python
#imagen2.shape nos devuelve alto ancho y colores de la imagen
alto,ancho,colores=imagen2.shape
```

```python
#Vamos a calcular las posiciones superior izquierda e inferior derecha
superiorIzquierda=poscionMaxima
inferiorDerecha=(poscionMaxima[0]+ancho,poscionMaxima[1]+alto)
```

```python
#Dibujamos un rectangulo dentro de la imagen1 donde esta la imagen del perro
#cv2.rectangle(imagen,superiorIzquierda,inferiorDerecha,color=(255,0,0),color)
cv2.rectangle(imagen,superiorIzquierda,inferiorDerecha,color=(255,0,0),thickness=8)
plt.imshow(imagen)
```
