```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#https://stackoverflow.com/questions/24564889/opencv-python-not-opening-images-with-imread
ajedrez=cv2.imread("D:\\CursoOpenCV\\Imagenes\\tablero-ajedrez.png")
ajedrezGris=cv2.cvtColor(ajedrez,cv2.COLOR_BGR2GRAY)
plt.imshow(ajedrezGris, cmap="gray")
```

```python
#Vamos a identificar las esquinas de cada color
#El metodo cv2.goodFeaturesToTrack(imagen,numeroDeEsquinas,parametro1,parametro2)
#Devuelve la posicion de todas las esquinas encontradas en una imagen
esquinas=cv2.goodFeaturesToTrack(ajedrezGris,100,0.01,10)
```

```python
#Convertimos los valor de esquinas a enteros
esquinas=np.int0(esquinas)
```

```python
#Para cada esquina 
for i in esquinas:
    #Metodo ravel para escojer de una a una
    x,y=i.ravel()
    #Dibujamos un circulo para cada esquina
    #cv2.circle(imagen,center=(250,250),radio,color=(0,255,0),grosor)
    cv2.circle(ajedrez,(x,y),4,(255,0,0),8)
    
```

```python
#Mostramos
plt.imshow(ajedrez)
```
